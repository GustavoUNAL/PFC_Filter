```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import random

# Configuración inicial de la figura y el gráfico
fig, ax = plt.subplots()
xdata, ydata = [], []
ln, = plt.plot([], [], 'r-', animated=True)

# Función de inicialización
def init():
    ax.set_xlim(0, 100)  # Ajusta el límite según tus datos
    ax.set_ylim(-1, 1)   # Ajusta el límite según tus datos
    return ln,

# Función de actualización
def update(frame):
    xdata.append(frame)
    # Genera un nuevo valor de la señal
    ydata.append(np.sin(frame * 0.1) + random.uniform(-0.1, 0.1))
    
    # Si la lista de datos es muy larga, recorta los primeros elementos
    if len(xdata) > 100:
        xdata.pop(0)
        ydata.pop(0)
    
    ln.set_data(xdata, ydata)
    ax.set_xlim(max(0, frame - 100), frame + 10)
    return ln,

# Configura la animación
ani = FuncAnimation(fig, update, frames=np.arange(0, 200, 0.5), init_func=init, blit=True, interval=100)  # Intervalo en milisegundos

plt.show()
```
